```python
# Seleccionar solo las columnas numéricas
columnas_numericas = datos.select_dtypes(include=[np.number]).columns
print(f"Columnas numéricas seleccionadas: {list(columnas_numericas)}")


```

```python
# Calcular el número de columnas y filas necesarias automáticamente
n = len(columnas_numericas)
ncols = 3  # Fijamos 3 columnas
nrows = (n + 2) // 3  # Esto asegura que se calculen las filas correctamente

print(f"Columnas numéricas seleccionadas: {list(columnas_numericas)}")
print(f"Subplots necesarios: {nrows} filas x {ncols} columnas")
```

```python
# Configurar el número de filas y columnas para los subplots
n = len(columnas_numericas)
nrows = 3
ncols = min(n, 3)

# Crear la figura y subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5*nrows))
fig.suptitle('Distribución de Variables Numéricas', fontsize=16)

# Asegurar que los ejes estén en un formato compatible
axes = axes.flatten() if n > 3 else [axes]

# Crear un histograma para cada variable numérica
for i, col in enumerate(columnas_numericas):
    ax = axes[i]
    # Use 'datos' instead of 'calidad_aire' to access your DataFrame
    datos[col].hist(ax=ax, bins=50, edgecolor='fuchsia', color='black')
    ax.set_title(f'Distribución de {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frecuencia')

# Ocultar cualquier subplot vacío
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
